```
from rdflib import Graph, Namespace, Literal, URIRef
g1 = Graph()
g2 = Graph()
g1.parse(github_storage+"resources/data01.rdf", format="xml")
g2.parse(github_storage+"resources/data02.rdf", format="xml")
```

Tarea: lista todos los elementos de la clase Person en el primer grafo (data01.rdf) y completa los campos (given name, family name y email) que puedan faltar con los datos del segundo grafo (data02.rdf). Puedes usar consultas SPARQL o iterar el grafo, o ambas cosas.

```
#Primer grafo
from rdflib import RDF, RDFS, XSD
ns = Namespace("http://data.org#")
vcard= Namespace("http://www.w3.org/2001/vcard-rdf/3.0#")
print("Elementos de la clase Person:")
for s,p,o in g1.triples((None, RDF.type, ns.Person )):
  print(s,p,o)

print("Completar los datos del segundo grafo:")

g2.add((ns.SaraJones,vcard.FN,Literal("Sara Jones",datatype=XSD.string)))
g2.add((ns.JohnSmith,vcard.FN,Literal("John Smith",datatype=XSD.string)))
g2.add((ns.JohnDoe,vcard.FN,Literal("John Doe",datatype=XSD.string)))
g2.add((ns.HarryPotter,vcard.Family,Literal("Potter",datatype=XSD.string)))

for s,p,o in g2:
  print(s,p,o)
```
