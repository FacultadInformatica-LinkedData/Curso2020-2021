Leemos el fichero RDF de la forma que lo hemos venido haciendo

```
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('ns', Namespace("http://somewhere#"), override=False)
g.namespace_manager.bind('vcard', Namespace("http://www.w3.org/2001/vcard-rdf/3.0#"), override=False)
g.parse(github_storage+"/resources/example6.rdf", format="xml")
```

**TASK 7.1: List all subclasses of "Person" with RDFLib and SPARQL**

```
from rdflib.plugins.sparql import prepareQuery
ns = Namespace("http://somewhere#")

q0 = prepareQuery('''
  SELECT ?s
  WHERE { 
    ?s rdfs:subClassOf ns:Person. 
  }
  ''',
  initNs={"rdfs":RDFS,"ns":ns}
)
print("Listing with RDFLib... \n")
for s,p,o in g.triples((None,RDFS.subClassOf,ns.Person)):
  print(s,p,o)
  
print("\n\nListing with SPARQL... \n")
for ret in g.query(q0):
  print(ret)
```

**TASK 7.2: List all individuals of "Person" with RDFLib and SPARQL (remember the subClasses)**


```
q0 = prepareQuery('''
  SELECT ?x
  WHERE { 
    {   
      ?x rdf:type ns:Person.   
    }
     UNION 
    {
      ?subclass rdfs:subClassOf ns:Person.
      ?x rdf:type ?subclass
    }
  }
  ''',
  initNs={"rdfs":RDFS,"rdf":RDF,"ns":ns}
)
print("Listing with RDFLib... \n")
for x,y,z in g.triples((None, RDF.type, ns.Person)):
  print(x)
for x,y,z in g.triples((None, RDFS.subClassOf, ns.Person)):
  for a,b,c in g.triples((None, RDF.type, x)):
    print(a)

print("\n\nListing with SPARQL... \n")
for aux in g.query(q0):
  print(aux)
```

**TASK 7.3: List all individuals of "Person" and all their properties including their class with RDFLib and SPARQL**


```
q0 = prepareQuery('''
  SELECT ?x ?y ?z
  WHERE 
  { 
    { 
      ?x rdf:type ns:Person.
      ?x ?y ?z
    } 
	UNION 
	{
      ?x rdf:type ?obj.
      ?x ?y ?z.
      ?obj rdfs:subClassOf ns:Person
    }
  }
  ''',
  initNs = { "rdfs": RDFS,"rdf": RDF, "ns": ns}
)
print("Listing with RDFLib... \n")
for x,z,y in g.triples((None, RDF.type, ns.Person)):
  for obj,p1,o1 in g.triples((x, None, None)):
      print(obj,p1,o1)

for x,z,y in g.triples((None, RDFS.subClassOf, ns.Person)):
  for obj,p1,o1 in g.triples((None, RDF.type, x)):
    for s2,p2,o2 in g.triples((obj, None, None)):
      print(s2,p2,o2)
      
print("\n\nListing with SPARQL... \n")
for aux in g.query(q0):
  print(aux)
```
