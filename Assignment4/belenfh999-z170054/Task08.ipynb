Tarea: lista todos los elementos de la clase Person en el primer grafo (data01.rdf) y completa los campos (given name, family name y email) que puedan faltar con los datos del segundo grafo (data02.rdf). Puedes usar consultas SPARQL o iterar el grafo, o ambas cosas.

**Task 1. List all elements from the Class Person in the first graph.**


```
from rdflib.namespace import RDF
base = Namespace("http://data.org#")
print("List of all elements from the class Person using RDFLib")
for s,p,o in g1.triples((None,RDF.type,base.Person)):
  for s0,p0,o0 in g1.triples((s,None,None)):
    print(s0,p0,o0)
print("List of all elements from the class Person using SPARQL")
qres = g1.query(
"""SELECT DISTINCT ?person ?props ?objs
WHERE {
?person a <http://data.org#Person>.
?person ?props ?objs.
}""")
for row in qres:
  print("%s %s %s" % row)
```

**Task 2. Complete the following fields: full name, given name, family name and email in the second graph.**

```
vcard= Namespace("http://www.w3.org/2001/vcard-rdf/3.0#")
g2.add((base.SaraJones,vcard.FN,Literal("Sara Jones",datatype=XSD.string)))
g2.add((base.JohnSmith,vcard.FN,Literal("John Smith",datatype=XSD.string)))
g2.add((base.JohnDoe,vcard.FN,Literal("John Doe",datatype=XSD.string)))
g2.add((base.HarryPotter,vcard.Family,Literal("Potter",datatype=XSD.string)))
for s,p,o in g2:
  print(s,p,o)
```
